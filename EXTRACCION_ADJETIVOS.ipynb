## CREAMOS DF_ESPAÑOL Y DF_INGLES

```python
import pandas as pd

# Creamos df español
data = {
    'frase': [
        "El perro marrón corre rápido.",
        "La casa grande está en la colina.",
        "Ella es una excelente pianista.",
        "El cielo azul es hermoso hoy.",
        "El restaurante italiano sirve deliciosa pizza.",
        "Mi hermana menor es muy inteligente."
    ]
}

df = pd.DataFrame(data)
```

## PRIMERA HERRAMIENTA

```python
# TRABAJAMOS CON df (NO FUNCIONA)
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag

# Descargar los recursos necesarios de NLTK si no lo has hecho antes
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

# Función para extraer adjetivos calificativos de una frase
def extract_adjectives(text):
    tokens = word_tokenize(text)  # Tokeniza la frase
    tagged = pos_tag(tokens)      # Etiqueta las palabras con su categoría gramatical
    adjectives = [word for word, pos in tagged if pos.startswith('JJ')]  # Filtra adjetivos
    return adjectives

# Aplica la función a cada fila del DataFrame y crea una nueva columna 'adjetivos'
df['adjetivos'] = df['frase'].apply(extract_adjectives)

# Muestra el DataFrame con la columna 'adjetivos'
print(df[['frase', 'adjetivos']])
```

```python
# TRABAJAMOS CON df ( FUNCIONA)
import spacy
import pandas as pd

# Cargar el modelo de SpaCy en español
nlp = spacy.load("es_core_news_sm")

# Función para extraer adjetivos calificativos de una frase utilizando SpaCy
def extract_adjectives_spacy(text):
    # Tokenizar la frase y etiquetar las partes del discurso con SpaCy
    doc = nlp(text)

    # Extraer adjetivos calificativos
    adjectives = [token.text for token in doc if token.pos_ == "ADJ"]

    return adjectives

# Aplicar la función a la columna 'frase' del DataFrame y crear una nueva columna 'adjetivos'
df['adjetivos'] = df['frase'].apply(extract_adjectives_spacy)

# Mostrar el DataFrame con la columna 'adjetivos'
print(df[['frase', 'adjetivos']])
```
