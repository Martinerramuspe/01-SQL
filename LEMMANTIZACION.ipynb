```python
import pandas as pd
import spacy
```

##02-LEMMATIZACION PALABRAS SUELTAS

```python
# Cargar el modelo de spacy para español
nlp = spacy.load('es_core_news_sm')
```

```python
# Crear un DataFrame con palabras español ( puede ser en ingles tambien )
data = {'palabras': ['corriendo', 'corrió', 'corredor', 'corre', 'corriendo']}
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.info()
```

```python
# Observamos lo que dice la primera palabra
df.loc[1]
```

```python
# Función para lematizar palabras
def lemmatize_words(text):
    doc = nlp(text)
    lemmatized_words = [token.lemma_ for token in doc]
    return ' '.join(lemmatized_words)
```

```python
# Aplicamos funcion, no es perfecta!!!
df["lemmatizado"]=df["palabras"].apply(lemmatize_words)
```

```python
#Funciona bien, y puede fallar con algunas palabras!
df.head()
```

##03-TRATAMIENTO DE LISTA DE PALABRAS POR CONTENEDOR.

```python
# Crear un DataFrame con una columna de cadenas separadas por comas
data = {'cadena': ['corrio, correra, corriendo', 'corriendo, correr, corredor']}
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
# Dividir la cadena en una lista de palabras y crear una nueva columna con las palabras separadas
df['palabras'] = df['cadena'].str.split(', ')
```

```python
df.head()
```

```python
df['lemmatizado'] = df['palabras'].apply(lambda x: [lemmatize_words(palabra) for palabra in x])
```
