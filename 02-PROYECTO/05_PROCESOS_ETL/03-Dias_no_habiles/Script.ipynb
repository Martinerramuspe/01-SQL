```python
import requests
from bs4 import BeautifulSoup as bs
import re
import pandas as pd
from datetime import datetime
```

```python
sopa= bs(x.content)
# Extraemos tabla
table = sopa.find("table", class_="table table-striped table-bordered table-hover table-condensed cols-2")
```

```python
rows_00= sopa.find("tbody")
```

```python
# Extramos todas las filas de la tabla
rows= rows_00.find_all("tr")
```

```python
# Elemento de fila 0, columna 0
F00=rows[0].find_all("td")[0]
F00.text
```

```python
# Elemento de fila 0, columna 1
F01=rows[0].find_all("td")[1]
F01.text
```

```python
# Extraemos y volcamos en lista
Fecha=[]
Descripcion=[]
for x in rows [0:42]: # va de 0 a 41
  Fecha.append(x.find_all("td")[0].text)
  Descripcion.append(x.find_all("td")[1].text)
```

```python
data = {'Fecha': Fecha,'Descripcion': Descripcion}
df= pd.DataFrame(data)
```

```python
# Creamos funcion
def eliminar_elementos(cadena):
    return re.sub(r'\n$', '', cadena)
```

```python
# Aplicamos funcion para eliminar anomalia.
df['Fecha'] = df['Fecha'].apply(eliminar_elementos)
```

```python
# Como podemos ver, muchas de los datos de la columna fecha tienen un patron claro, pero hay algun con ciertas particularidades que debemos tratar
df.head(8)
```

```python
def identificar_dos_numeros(fecha_str):
    numeros_encontrados = sum(c.isdigit() for c in fecha_str)
    return numeros_encontrados > 2
```

```python
df['anomalo'] = df['Fecha'].apply(identificar_dos_numeros)
```

```python
df
```

```python
df_filtrado = df[df['anomalo'] == True]
```

```python
df_filtrado
```

```python
# Función para transformar la cadena de fecha a una lista de dos números y el mes
def transformar_fecha(cadena):
    numeros = re.findall(r'\d+', cadena)
    mes = re.search(r'(\b\w+\b)$', cadena).group(0)

    # Asegurar que haya al menos dos números
    if len(numeros) == 1:
        numeros.append(numeros[0])

    return [int(num) for num in numeros] + [mes]

# Función para expandir las filas según el intervalo de fechas con información completa de la fecha
def expandir_filas(fila):
    inicio, fin, mes = transformar_fecha(fila['Fecha'])
    fechas_expandidas = [(f"{i} {mes}", fila['Descripcion'], fila['anomalo']) for i in range(inicio, fin + 1)]
    return fechas_expandidas
```

```python
filas_expandidas = df_filtrado.apply(expandir_filas, axis=1).explode()
```

```python
df_expandido = pd.DataFrame(list(filas_expandidas), columns=['Fecha', 'Descripcion', 'anomalo'])
```
